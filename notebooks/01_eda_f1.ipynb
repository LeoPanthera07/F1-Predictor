# 01 – EDA: F1 Result Predictor

## 1. Data loading & merging

- Read `races.csv`, `circuits.csv`, `drivers.csv`, `results.csv`, `qualifying.csv`.
- Merge:
  - `results` + `races` → race metadata per result.
  - `results` + `drivers` → driver metadata per result.
  - `results` + `qualifying` → grid vs qualifying position.

## 2. Basic checks

- Missing values in `grid`, `position`, `status`.
- DNF / retirements count per driver / circuit.
- Distribution of:
  - `grid` vs `position`
  - `positions_gained` (grid − position)

## 3. Visualizations

- Scatter: grid vs finish position colored by circuit.
- Histograms:
  - Points per race
  - DNF rate per driver
- Boxplots:
  - Finish position by driver, by constructor
- Time‑series:
  - Driver avg finish over seasons

## 4. Target analysis

- Define target:
  - Option A: regression on `position` (float, 1–20).
  - Option B: classification into buckets (e.g., 1–3, 4–6, 7–10, 11–20).
- Check how easy grid is to predict the finish (e.g., correlation grid–finish per circuit).

## 5. Feature‑engineering ideas (to implement in code cells)

- Driver‑level:
  - `driver_avg_finish` (rolling / seasonal)
  - `driver_dnf_rate`
  - `driver_age` at race
- Circuit‑level:
  - `circuit_length`
  - `is_street_circuit` (boolean)
- Driver‑circuit history:
  - `driver_circuit_avg_finish`
  - `driver_circuit_race_count`
- Qualifying‑derived:
  - `qualified_position` vs `grid`
  - `qualified_to_grid_delta`
- Weather‑style (if available):
  - `weather` / `track_temp` as ordinal or categorical
